The main topic I am interested in is the amount of Nintendo games sold per console.
My questions are related to how many Nintendo games of each genre were sold given the following three Nintendo consoles:
The N64 (Nintendo 64)
The NGC (Nintendo Game Cube)
The Wii (Nintendo Wii)

This following graph is a heat map for what Nintendo games genres released on the N64. The reason I chose a heatmap is due to the sheer amount of diverse games that came out during that time. It had many important games in each genre such as, Mario Tennis and Golf for sports, Super Mario 64 for platforming, and Mario Kart 64 for racing, amongst others.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Creating the Data Frame
df = pd.read_csv(r"C:\Users\adamt\Downloads\data.csv")

#Filter it to make sure it is only N64 games
df_n64 = df[(df['Platform'] == 'N64') & (df['Publisher'] == 'Nintendo')]

# Pivot table: index = Genre, columns = Game Name (or leave aggregated), values = Sales
#Summing up sales per value
sales_by_genre = df_n64.groupby('Genre')['Global_Sales'].sum().reset_index()

#Making a simple Heatmap: genres on y-axis, sales as color
plt.figure(figsize=(8,6))
sns.heatmap(sales_by_genre.set_index('Genre').T, annot=True, fmt=".1f", cmap="YlGnBu")
plt.title("Total Sales of Nintendo N64 Games per Genre")
plt.xlabel("Genre")
plt.ylabel("Global Sales (millions)")
plt.show()
```

The graph showed me that platformers and racing games were the best selling genres on the console, which is not surpirsing to me. With every main stay 2d platformer making their debute in 3d on this console such as Mario 64, Donkey Kong 64, and Kirby and the Crystal Shards. This console also saw some newer franchises from third party developers release during the time such as, Banjo Kazooie, Banjo Tooie, and Conkers Bad Furry Day. Racing also had many new entires, both sequals and new games. These included, Wave Race 64, Excite Bike 64, Mario Kart 64, and F-Zero X.

```python
import pandas as pd
import matplotlib.pyplot as plt

#Creating the Data Frame
df = pd.read_csv(r"C:\Users\adamt\Downloads\data.csv")

#Filter it to make sure it is only Gamecube games
df_gc = df[df['Platform'] == 'GC']

#Find the sales per genre
genre_sales = df_gc.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False)

#Plotting the bar graph
plt.figure(figsize=(10,6))
genre_sales.plot(kind='bar', color='cornflowerblue')

plt.xlabel("Genre")
plt.ylabel("Total Sales (millions)")
plt.title("Nintendo GameCube: Total Sales per Genre")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

The Game Cube did surprise me. However, the platform data does make sense, as this console was also home to new entires on the Mario, Luigi's Mansion, and a lot of Sonic Titles made their way over to the Game Cube in package releases.

```python
import pandas as pd
import matplotlib.pyplot as plt

#Creating the Data Frame
df = pd.read_csv(r"C:\Users\adamt\Downloads\data.csv")

#Filter it to make sure it is only Wii games
df_wii = df[df['Platform'] == 'Wii']

#Find the sales per genre
genre_sales = df_wii.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False)

#Plotting the bar graph
plt.figure(figsize=(10,6))
genre_sales.plot(kind='bar', color='mediumseagreen')

plt.xlabel("Genre")
plt.ylabel("Total Sales (millions)")
plt.title("Nintendo Wii: Total Sales per Genre")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

The Wii's graph did not surprise me as both Wii Sports and Wii Sports Resort were pack in titles for the Wii on release. Not to mention that this console saw the most popular entries in the Mario Sports series with, Mario Strikers Charged and Mario Super Sluggers. The Misc category is as high as it is, due to the amount of random games that Nintendo themselves pushed for, such as Wii Music, Wii Fit, and Wii Play amongst other one off Wii titles that showed off the Wii remote's motion controls.
